```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete["school_name"].value_counts().count()
total_students = school_data_complete["Student ID"].value_counts().count()
total_budget = []

for i in school_data_complete["budget"]:
    if i not in total_budget:
        total_budget.append(i)
        
avg_math = round(school_data_complete["math_score"].mean(),2)
avg_read = round(school_data_complete["reading_score"].mean(),2)
#Math score of 70 or more
pass_math = round((school_data_complete[school_data_complete["math_score"]>=70]["math_score"].count()/total_students)*100,2)
#Reading score of 70 or more
pass_read = round((school_data_complete[school_data_complete["reading_score"]>=70]["reading_score"].count()/total_students)*100,2)
overall_pass = round((pass_math+pass_read)/2,2)

school_data_df = pd.DataFrame({
    "Total Schools" : [total_schools],
    "Total Students" : [total_students],
    "Total Budget" : sum(total_budget),
    "Average Math Score" : [avg_math],
    "Average Reading Score" : [avg_read],
    "% Passing Math" : [pass_math],
    "% Passing Reading" : [pass_read],
    "% Overall Passing Rate" : [overall_pass]
})

school_data_df["Total Budget"] = school_data_df["Total Budget"].astype(float).map("${:}".format)
for i in school_data_df.columns[5:]:
    school_data_df[i] = school_data_df[i].astype(float).map("{:}%".format)
school_data_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Groupby School Name
school_data_groupby_schoolname = school_data_complete.groupby(["school_name"])

school_type = school_data_complete.drop_duplicates("school_name")[["school_name","type"]].sort_values(by = ["school_name"],ascending = True).set_index("school_name")["type"]
school_total_students = school_data_groupby_schoolname["Student ID"].count()
school_budget = school_data_groupby_schoolname["budget"].mean()
per_student_budget = round(school_budget/school_total_students,2)
avg_math_score = round(school_data_groupby_schoolname["math_score"].mean(),2)
avg_reading_score = round(school_data_groupby_schoolname["reading_score"].mean(),2)
pass_math_ = round((school_data_complete[school_data_complete["math_score"]>=70][["school_name","math_score"]].groupby("school_name")["math_score"].count()/school_total_students)*100,2)
pass_reading_ = round((school_data_complete[school_data_complete["reading_score"]>=70][["school_name","reading_score"]].groupby("school_name")["reading_score"].count()/school_total_students)*100,2)
overall_pass = round((pass_math_+pass_reading_)/2,2)

school_performance_df = pd.DataFrame({
    "School Type" : school_type,
    "Total Students" : school_total_students,
    "Total School Budget" : school_budget,
    "Per Student Budget" : per_student_budget,
    "Average Math Score" : avg_math_score,
    "Average Reading Score" : avg_reading_score,
    "% Passing Math" : pass_math_,
    "% Passing Reading" : pass_reading_,
    "% Overall Passing Rate" : overall_pass   
})

for i in school_performance_df.columns[2:4]:
    school_performance_df[i] = school_performance_df[i].astype(float).map("${:}".format)
for i in school_performance_df.columns[6:]:
    school_performance_df[i] = school_performance_df[i].astype(float).map("{:}%".format)
school_performance_df

school_performance_df.sort_values(by = "% Overall Passing Rate",ascending=False).head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Call out to original DataFrame to prevent formatting issues
school_performance_df = pd.DataFrame({
    "School Type" : school_type,
    "Total Students" : school_total_students,
    "Total School Budget" : school_budget,
    "Per Student Budget" : per_student_budget,
    "Average Math Score" : avg_math_score,
    "Average Reading Score" : avg_reading_score,
    "% Passing Math" : pass_math_,
    "% Passing Reading" : pass_reading_,
    "% Overall Passing Rate" : overall_pass   
})
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
budget_student_groupby = school_performance_df.groupby(pd.cut(school_performance_df["Per Student Budget"], spending_bins, labels=group_names))
budget_math = round(budget_student_groupby["Average Math Score"].mean(),2)
budget_reading = round(budget_student_groupby["Average Reading Score"].mean(),2)
pass_math_budget = round(budget_student_groupby["% Passing Math"].mean(),2)
pass_read_budget = round(budget_student_groupby["% Passing Reading"].mean(),2)
overall_pass_budget = round((pass_math_budget + pass_read_budget) / 2,2)

budget_student_df = pd.DataFrame({
    "Average Math Score" : budget_math,
    "Average Reading Score" : budget_reading,
    "% Passing Math" : pass_math_budget,
    "% Passing Reading" : pass_read_budget,
    "% Overall Passing Rate" : overall_pass_budget
})

for i in budget_student_df.columns[2:]:
    budget_student_df[i] = budget_student_df[i].astype(float).map("{:}%".format)
budget_student_df
```

```python
#Alternative way to solve problem using aggregate
budget_student_df2 = school_performance_df.groupby(pd.cut(school_performance_df["Per Student Budget"], spending_bins, labels = group_names)).aggregate({
    "Average Math Score" : 'mean',
    "Average Reading Score" : 'mean',
    "% Passing Math" : 'mean',
    "% Passing Reading" : 'mean'
})

budget_student_df2["% Overall Passing Rate"] = (budget_student_df2["% Passing Math"]+budget_student_df2["% Passing Reading"])/2

for i in budget_student_df2.columns:
    budget_student_df2[i] = round(budget_student_df2[i],2)

for i in budget_student_df2.columns[2:]:
    budget_student_df2[i] = budget_student_df2[i].astype(float).map("{:}%".format)

budget_student_df2
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_size_groupby = school_data_complete.groupby(pd.cut(school_data_complete["size"], size_bins, labels = group_names))

school_size_df = school_size_groupby.aggregate({
    'math_score' : 'mean',
    'reading_score' : 'mean'
}).rename(columns={'math_score':'Average Math Score','reading_score':'Average Reading Score'})

school_size_count = school_size_groupby.aggregate({
    'math_score' : 'count',
    'reading_score' : 'count'
})
```

```python
math_pass_df = school_data_complete[[val>=70 for val in school_data_complete["math_score"].values]]
math_pass_size_df = math_pass_df.groupby(pd.cut(math_pass_df["size"], size_bins, labels = group_names)).aggregate({
    "math_score" : 'count'
})

read_pass_df = school_data_complete[[val>=70 for val in school_data_complete["reading_score"].values]]
read_pass_size_df = read_pass_df.groupby(pd.cut(read_pass_df["size"], size_bins, labels = group_names)).aggregate({
    "reading_score" : 'count'
})

school_size_df["% Passing Math"] = (math_pass_size_df["math_score"] / school_size_count["math_score"]) * 100
school_size_df["% Passing Reading"] = (read_pass_size_df["reading_score"] / school_size_count["reading_score"]) * 100
school_size_df["% Overall Passing Rate"] = (school_size_df["% Passing Math"] + school_size_df["% Passing Reading"]) / 2

for i in school_size_df.columns:
    school_size_df[i] = round(school_size_df[i],2)
    
for i in school_size_df.columns[2:]:
    school_size_df[i] = school_size_df[i].astype(float).map("{:}%".format)
    
school_size_df
```
