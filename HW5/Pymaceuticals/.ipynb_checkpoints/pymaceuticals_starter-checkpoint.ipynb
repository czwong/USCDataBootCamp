```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drug_trial_df = pd.merge(mouse_drug_data, clinical_trial_data, how = 'left', on = 'Mouse ID')

# Display the data table for preview
drug_trial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = drug_trial_df.groupby(['Drug',  'Timepoint'])['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tumor_df = pd.DataFrame(mean_tumor)

# Preview DataFrame
mean_tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
ste_tumor = drug_trial_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
ste_tumor_df = pd.DataFrame(ste_tumor)

# Preview DataFrame
ste_tumor_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_df = mean_tumor_df.reset_index(level = ['Drug','Timepoint']).pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
mean_tumor_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
ste_tumor_df = ste_tumor_df.reset_index(level = ['Drug','Timepoint']).pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
ste_tumor_df.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

ax.errorbar(mean_tumor_df.index, mean_tumor_df['Capomulin'], ste_tumor_df['Capomulin'], fmt = 'o', color = 'b', linestyle = ':', label = 'Capomulin')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Ceftamin'], ste_tumor_df['Ceftamin'], fmt = 'v', color = 'orange', linestyle = ':', label = 'Ceftamin')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Infubinol'], ste_tumor_df['Infubinol'], fmt = 'h', color = 'g' ,linestyle = ':', label = 'Infubinol')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Ketapril'], ste_tumor_df['Ketapril'], fmt = 's', color = 'r' ,linestyle = ':', label = 'Ketapril')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Naftisol'], ste_tumor_df['Naftisol'], fmt = 'p', color = 'purple' ,linestyle = ':', label = 'Naftisol')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Placebo'], ste_tumor_df['Placebo'], fmt = 'x', color = 'brown' ,linestyle = ':', label = 'Placebo')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Propriva'], ste_tumor_df['Propriva'], fmt = '+', color = 'pink' ,linestyle = ':', label = 'Propriva')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Ramicane'], ste_tumor_df['Ramicane'], fmt = 'D', color = 'grey' ,linestyle = ':', label = 'Ramicane')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Stelasyn'], ste_tumor_df['Stelasyn'], fmt = '8', color = 'gold' ,linestyle = ':', label = 'Stelasyn')
ax.errorbar(mean_tumor_df.index, mean_tumor_df['Zoniferol'], ste_tumor_df['Zoniferol'], fmt = '2', color = 'cyan' ,linestyle = ':', label = 'Zoniferol')

ax.grid(linewidth = '1')
plt.title('Tumor Response Treatment', Fontsize = 20)
plt.xlabel('Time (Days)', Fontsize = 10)
plt.ylabel('Tumor Volume (mm3)', Fontsize = 10)
ax.set_ylim(30, 75)
ax.legend(loc=(1.05,0.15))

# Save the Figure
plt.savefig("../Images/Tumor Response Treatment.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic = drug_trial_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
mean_metastatic_df = pd.DataFrame({'Metastatic Sites' : mean_metastatic})

# Preview DataFrame
mean_metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint
ste_metastatic = drug_trial_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
ste_metastatic_df = pd.DataFrame({'Metastatic Sites' : ste_metastatic})

# Preview DataFrame
ste_metastatic_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_metastatic_df = mean_metastatic_df.reset_index(level = ['Drug', 'Timepoint']).pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')

# Preview that Reformatting worked
mean_metastatic_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
ste_metastatic_df = ste_metastatic_df.reset_index(level = ['Drug', 'Timepoint']).pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')

# Preview that Reformatting worked
ste_metastatic_df.head()
```

```python
# Generate the Plot (with Error Bars)
markers = ['o', 'v', 'h', 's', 'p', 'x', '+', 'D', '8', '2']

fig, ax = plt.subplots()

for i in mean_metastatic_df.columns:
    ax.errorbar(mean_metastatic_df.index, mean_metastatic_df[i], ste_metastatic_df[i], fmt = markers[list(mean_metastatic_df.columns).index(i)], linestyle = ':', label = i)

ax.grid(linewidth = '1')
plt.xlabel("Treatment Duration(Days)", Fontsize = 10)
plt.ylabel('Met. Sites', Fontsize = 10)
plt.title('Metastatic Spread During Treatment', Fontsize = 20)
ax.legend(loc = (1.05,0.15))

# Save the Figure
plt.savefig("../Images/Metastatic Spread During Treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = drug_trial_df.groupby(['Drug', 'Timepoint'])['Mouse ID'].count()

# Convert to DataFrame
mice_count_df = pd.DataFrame({'Mice Count' : mice_count})

# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_df = mice_count_df.reset_index(level = ['Drug', 'Timepoint']).pivot(index = 'Timepoint', columns = 'Drug', values = 'Mice Count')

# Preview the Data Frame
mice_count_df.head()
```

```python
# Calculate the percent survival rate
for i in mice_count_df:
    mice_count_df[i] = mice_count_df[i]/mice_count_df[i][0] * 100

# Display the data to confirm
mice_count_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
markers = ['o', 'v', 'h', 's', 'p', 'x', '+', 'D', '8', '2']
    
fig, ax = plt.subplots()

for i in mice_count_df:
    ax.errorbar(mice_count_df.index, mice_count_df[i], marker = markers[list(mice_count_df.columns).index(i)], linestyle = ':', label = i)

ax.grid(linewidth = '1')
plt.xlabel('Time (Days)', Fontsize = 10)
plt.ylabel('Survival Rate (%)', Fontsize = 10)
plt.title('Survival During Treatment', Fontsize = 20)
ax.legend(loc = (1.05, 0.15))

# Save the Figure
plt.savefig('../Images/Survival During Treatment.png')

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
percent_change = []
for drug in mean_tumor_df.columns:
    percent_change.append((mean_tumor_df[drug].iloc[-1] - mean_tumor_df[drug].iloc[0]) / mean_tumor_df[drug].iloc[0] * 100)

# Display the data to confirm
percent_change_df = pd.DataFrame({
    'Drugs' : mean_tumor_df.columns,
    'Percent Change' : percent_change
})

percent_change_df.head()
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change_tuple = tuple(percent_change_df['Percent Change'])

# Splice the data between passing and failing drugs
passing_drugs = []
failing_drugs = []

for i in percent_change_tuple:
    if i < 0:
        failing_drugs.append(percent_change_df['Drugs'].iloc[percent_change_tuple.index(i)])
    else:
        passing_drugs.append(percent_change_df['Drugs'].iloc[percent_change_tuple.index(i)])


# Orient widths. Add labels, tick marks, etc.
x_axis = np.arange(0,10,1)

fig, ax = plt.subplots(figsize = (20,10))
ax.grid(linewidth = '1')
bar = ax.bar(x_axis, percent_change_tuple, align = 'edge', width = -1)

for i in range(len(percent_change_tuple)):
    if percent_change_df['Drugs'].iloc[i] in passing_drugs:
        color = 'r'
        bar[i].set_color(color)
    else:
        color = 'g'
        bar[i].set_color(color)

ax.set_xlim(-2, max(x_axis)+1)
ax.set_ylim(min(percent_change_tuple)-5, max(percent_change_tuple)+5)
plt.xticks(x_axis, list(percent_change_df['Drugs']), rotation = 0, Fontsize = 15)
plt.ylabel('% Tumor Volume Change', Fontsize = 15)
plt.title('Tumor Change Over 45 Day Treatment', Fontsize = 20)

# Use functions to label the percentages of changes
rects = ax.patches

for rect in rects:
    height = rect.get_height()
    width = (((rect.get_x()-1) + rect.get_x()) / 2)
    if height > 0:
        ax.text(width, height, f'{round(height,2)}%', Fontsize = 20, ha = 'center', va = 'bottom')
    else:
         ax.text(width, height, f'{round(height,2)}%', Fontsize = 20, ha = 'center', va = 'top')

# Call functions to implement the function calls


# Save the Figure
plt.savefig('../Images/Tumor Change Over 45 Day Treatment.png')

# Show the Figure
plt.show()
```
