```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
beer_df = pd.read_csv("open_units.csv", header = None, 
                      names = ["Product", "Brand", "Category", "Style", "Quantity", "Quantity Units", "Volume", "Package", "ABV", "Units of Alcohol", "Units (4 Decimal Places)", "Units per 100 ml"])
beer_df.head()
```

```python
abv_median = beer_df["ABV"].sum() / beer_df["ABV"].count()
abv_median
```

```python
group_name = ["Lower ABV", "Higher ABV"]
bin_size = [0, abv_median, max(beer_df["ABV"])]

beer_df["ABV Level"] = pd.cut(beer_df["ABV"], bin_size, labels = group_name)
```
